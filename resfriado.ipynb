#### Clasificacion

```python
import pandas as pd 

#Dataset codificado 
data = {
    "Fiebre_alta": [1,1,0,0,1],
    "Dolor_muscular":[1,1,0,1,0],
    "Congestion_nasal":[0,1,1,0,1],
    "Diagnostico":[1,1,0,0,1] #1=Gripe, 0=Resfriado
}

df = pd.DataFrame(data)
#Se divide en dos partes: caracteristicas (x) y variable objetivo (y)
X = df.drop("Diagnostico", axis=1) #Caracteristicas .drop elimina la columna "Diagnostico", axis=1 indica que es una columna 
y = df["Diagnostico"]

print("Dataset original")
print(df)
```

#### 3: Entrenar el modelo 

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

#Crear y entrenar el modelo 
model = DecisionTreeClassifier(criterion="gini",max_depth=3,random_state=42)
model.fit(X,y) #Necesita el metodo fit para entrenar el modelo con los datos de entrenamiento (X,y)

#Visualizar el arbol 
plt.figure(figsize=(12,3))
plot_tree(model,
          feature_names=X.columns,
          class_names=["Resfriado","Gripe"],
          filled = True,
          rounded=True)
plt.show()
```

```python
'''Fiebre alta «= 0.5.
• Condición de división: Si "Fiebre alta" ess 0.5 (No), sigue por la rama izquierda (True). Si no, por lo derecha (False),
Codificación: 0 = No, 1 = Sí (porque los datos se codificaron numéricamente).

gin1 = 0.48
impureza del nodo: Mide cuán mezcladas están las clases. Cercano a 0.5 indica máxima impureza (50% Gripe, 50% Resfriado).

samples = S:
• Número de muestras en este nodo (todal las 5 muestras del dataset).

value = [2, 3]:
Distribución de clases: 2 'Resfriado' (clase 0) y 3 "Gripe" (clase 1).

Class = Gripe:
• Clase mayoritaria: Como hay más "Gripe" (3 vs 2), este nodo predeciría "Gripe' si fuera una roja.'''
```

#### 4 entrenar el modelo 

```python
#Prececir un nuevo paciente: Fiebre=No(0), Dolor=No(0), Congestión=Sí(1)
nuevo_paciente=pd.DataFrame([[0,0,1]],columns=X.columns)
prediccion= model.predict(nuevo_paciente)
if prediccion[0]==1:
    print("Diagnóstico: Gripe")
else:
    print("Diagnóstico: Resfriado")
```

# Regresion

```python
import pandas as pd 

#Dataset codificado 
data = {
    "Fiebre_alta": [1,1,0,0,1],
    "Dolor_muscular":[1,1,0,1,0],
    "Congestion_nasal":[0,1,1,0,1],
    "Duracion":[7,10,3,5,8]
}

df = pd.DataFrame(data)
#Se divide en dos partes: caracteristicas (x) y variable objetivo (y)
X = df.drop("Duracion", axis=1) #Caracteristicas .drop elimina la columna "Duracion", axis=1 indica que es una columna 
y = df["Duracion"]

print("Dataset original")
print(df)
```

## Entrear el modelo 

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt

#Crear y entrenar el modelo 
model = DecisionTreeRegressor(criterion='squared_error', # Usa MSE para regresion 
                               max_depth=2, #Limitamos la profunidad para simplificar 
                               random_state=42)
model.fit(X,y)

#Visualizar el arbol 
plt.figure(figsize=(12,6))
plot_tree(model,
          feature_names=X.columns,
          filled = True,
          rounded=True,
          precision=1)
plt.show()
```

## 4 Hacer prediciones

```python
# Predecir para un nuevo paciente: Fiebre=1,Dolor=0,Congestion=1
nuevo_paciente=pd.DataFrame([[1,0,1]],columns=X.columns)
prediccion = model.predict(nuevo_paciente)
print(f"Duracion predicha: {prediccion[0]:.1f} dias ")
```
